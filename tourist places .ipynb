```python
df=pd.read_csv('Top Indian Places to Visit.csv')
df
```

```python
df=df.drop(columns=['Unnamed: 0','Name','Zone','State','City','time needed to visit in hrs','Establishment Year','Airport with 50km Radius','Weekly Off','Significance','DSLR Allowed','Best Time to visit'])
df
```

```python
df=df.dropna()
df
```

```python
df.drop(df[df['Entrance Fee in INR'] == 0].index, inplace=True)
df
```

```python
df=pd.get_dummies(df,dtype=int)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop(columns=['Entrance Fee in INR'])
Y= df['Entrance Fee in INR']
X
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.60)
model = LinearRegression()
model.fit(X_train,Y_train)
ans = model.predict(X_test)
print(ans)
```

```python
model.score(X_train,Y_train)
```
